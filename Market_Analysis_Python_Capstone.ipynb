=======================================================================================Task 2 

The cost column in the dataframe “Project” has some missing values. Your task is to 
compute these missing values. Replace the missing values by running average. You 
should use the “For” loop for this task. 

```python
# Check how many missing values are in Cost column
project_df["Cost"].isnull().sum()
```

```python
# Create variables to store running total and count
running_sum = 0
count = 0

# Loop through each row index in project_df
for i in range(len(project_df)):

    # Check if value is NOT missing
    if not pd.isnull(project_df.loc[i, "Cost"]):
        
        # Add value to running sum
        running_sum += project_df.loc[i, "Cost"]
        
        # Increase count
        count += 1
    
    # If value IS missing
    else:
        
        # Calculate running average
        running_avg = running_sum / count
        
        # Replace missing value with running average
        project_df.loc[i, "Cost"] = running_avg
```

=======================================================================================Task 4

Join all three dataframes in one single dataframe. Name it “Final”

```python
# Load employee data
employee_df = pd.read_csv("employee.csv")

# Load seniority data
seniority_df = pd.read_csv("seniority.csv")

# Load project data
project_df = pd.read_csv("project.csv")
```

```python
# Merge employee and seniority data on ID
emp_seniority = pd.merge(employee_df, seniority_df, on="ID")

# Display result
emp_seniority
```

```python
# Merge previous result with project data on ID
Final = pd.merge(emp_seniority, project_df, on="ID")

# Display Final DataFrame
Final
```

```python
# Save Final dataframe
Final.to_csv("Final.csv", index=False)
```

=======================================================================================Task 5

Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
cost only to employees who have finished the projects.

```python
# Create Bonus column

Final["Bonus"] = 0  # Initialize column with 0

# Loop through each row
for i in range(len(Final)):
    
    # If project status is Finished
    if Final.loc[i, "Status"] == "Finished":
        
        # Give 5% bonus
        Final.loc[i, "Bonus"] = Final.loc[i, "Cost"] * 0.05
```

```python
Final[["Project", "Cost", "Status", "Bonus"]]
```

```python
Final.to_csv("Final.csv", index=False)
```

=======================================================================================Task 6

Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4.

```python
# Loop through each row
for i in range(len(Final)):
    
    # If project status is Failed
    if Final.loc[i, "Status"] == "Failed":
        
        # Reduce designation level by 1
        Final.loc[i, "Designation Level"] += 1
```

```python
# Remove rows where designation level is greater than 4
Final = Final[Final["Designation Level"] <= 4]
```

```python
Final[["ID", "Project", "Status", "Designation Level"]]
```

```python
Final.to_csv("Final.csv", index=False)
```

=======================================================================================Task 7

Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# Loop through each row
for i in range(len(Final)):
    
    # If Gender is Male
    if Final.loc[i, "Gender"] == "M":
        Final.loc[i, "First Name"] = "Mr. " + Final.loc[i, "First Name"]
    
    # If Gender is Female
    elif Final.loc[i, "Gender"] == "F":
        Final.loc[i, "First Name"] = "Mrs. " + Final.loc[i, "First Name"]
```

```python
# Remove Gender column
Final = Final.drop("Gender", axis=1)
```

=======================================================================================Task 8

Promote designation level by 1 for the employees whose age is more than 29 years using
IF condition.

```python
# Loop through each row
for i in range(len(Final)):
    
    # If Age is greater than 29
    if Final.loc[i, "Age"] > 29:
        
        # Increase designation level by 1
        Final.loc[i, "Designation Level"] -= 1
```

=======================================================================================Task 9

Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
# Group by ID and First Name
TotalProjCost = Final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename Cost column to Total Cost
TotalProjCost = TotalProjCost.rename(columns={"Cost": "Total Cost"})

# Display the new DataFrame
TotalProjCost
```

```python
# Save TotalProjCost dataframe
TotalProjCost.to_csv("TotalProjCost.csv", index=False)
```
